```python
# Change a column name to be consistent with most others and to eliminate the white space
df = od.rename(columns = {'Sex ': 'Sex', 'Species ': 'Species', 'original order':'Original Order'})
df
```

```python
# check data type of the columns 
df.dtypes
```

It shows that in the age columns there are data like "60s", not only consisting of numbers, so it can't change to numeric for now. We can deal with that later.

```python
# check for missing values
df.isna().sum()
```

```python
# percentage of missing values in each column
df.isna().sum()/len(df)
```

It makes sense to eliminate the columns with many missing values (e.g. 30%).
I eliminate columns "Unnamed: 22" and "Unnamed: 23". Other columns such as "Age", "Time" and "Species" also have 
high percentage of missing values, but those are very interesting characteristics to know, so I keep them.

```python
# drop 2 columns with many NaN
df = df.drop(columns=['Unnamed: 22', 'Unnamed: 23'])
df
```

```python
# fill missing values in 'Age'  with zeros
df['Age'] = df['Age'].fillna(0)
df['Age']
```

```python
# check whether "Case Number" could be used as unique identifier
df['Case Number'].is_unique
```

```python
# Since it's not, check where / why not
df['Case Number'].value_counts()
```

```python
# check whether column 'Original Order' could be used as unique identifier
df['Original Order'].is_unique
```

```python
# check for data in column 'Name'
df['Name'].value_counts()
```

```python
# Since 'male', 'female', 'boy, '2 males', 'boat' are not names, replace them with NaN
invalid_names = ['male', 'female', 'boy', '2 males', 'boat']
df['Name'] = np.where(df['Name'].isin(invalid_names), 'np.nan', df['Name'])
df
```

```python
# check for variance of columns
df['Year'].describe()
```

```python
# Save the cleaner dataset to a new csv-file
df.to_csv('SharkAttack_clean.csv')
```
